```python
monthlyFee = float(input("Enter your monthly fee: "))
yearlyFee = monthlyFee * 12

print("Your yearly fee = ", yearlyFee)

if yearlyFee < 10000:
  print("Cashback: ", yearlyFee * 0.01, "Baht")
elif yearlyFee > 10000 and yearlyFee <= 30000:
  print("Cashback: ", yearlyFee * 0.05, "Baht")
elif yearlyFee > 30000:
  print("Cashback: ", yearlyFee * 0.1, "Baht")
else:
  print("Cannot calculate your cashback.")
```
