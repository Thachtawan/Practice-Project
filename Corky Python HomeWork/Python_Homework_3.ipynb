```python
menu = 0
balAmount = float(0)

while menu != 4:
  menu = int(input("Please enter menu (1-4): 1. Deposit 2. Withdraw 3. Check Balance 4. Exit: "))
  if menu == 1:
    depositAmount = float(input("Enter amount deposit THB: "))
    balAmount += depositAmount
  elif menu == 2:
    wdAmount = float(input("Enter amount withdraw THB: "))
    if wdAmount > balAmount:
      print("You can't withdraw with amount greater than balance. Please try again.")
    else:
      balAmount -= wdAmount
      print("You have ", balAmount, "THB in your account")
  elif menu == 3:
    print("Your balance is ", balAmount, "THB")
  elif menu == 4:
    print("Thank you and goodbye.")
  else:
    print("Wrong order please try again!")
```
