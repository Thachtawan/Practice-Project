```python
monthSal = float(input("Enter your salary here: "))
yearlySal = monthSal * 12

print("Your yearly salary =", yearlySal, " THB")

if yearlySal <= 150000:
  print("Your personal tax income = 0 THB")
elif yearlySal > 150000 and yearlySal <= 300000:
  print("Your personal tax income = ", (yearlySal - 150000) * 0.05, " THB")
elif yearlySal > 300000 and yearlySal <= 500000:
  print("Your personal tax income = ", ((yearlySal - 300000) * 0.1) + 150000 * 0.05, " THB")
elif yearlySal > 500000 and yearlySal <= 750000:
  print("Your personal tax income = ", ((yearlySal - 500000) * 0.15) + 200000 * 0.1 + 150000 * 0.05, " THB")
elif yearlySal > 750000 and yearlySal <= 1000000:
  print("Your personal tax income = ", ((yearlySal - 750000) * 0.2) + 250000 * 0.15 + 200000 * 0.1 + 150000 * 0.05, " THB")
elif yearlySal > 1000000 and yearlySal <= 2000000:
  print("Your personal tax income = ", ((yearlySal - 1000000) * 0.25) + 250000 * 0.2 + 250000 * 0.15 + 200000 * 0.1 + 150000 * 0.05, " THB")
elif yearlySal > 2000000 and yearlySal <= 5000000:
  print("Your personal tax income = ", ((yearlySal - 2000000) * 0.3) + 1000000 * 0.25 + 250000 * 0.2 + 250000 * 0.15 + 200000 * 0.1 + 150000 * 0.05, " THB")
else:
  print("Your personal tax income = ", ((yearlySal - 5000000) * 0.35) + 3000000 * 0.3 + 1000000 * 0.25 + 250000 * 0.2 + 250000 * 0.15 + 200000 * 0.1 + 150000 * 0.05, " THB")
```
