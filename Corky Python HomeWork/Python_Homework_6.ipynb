```python
import random

rdNumber = random.randint(1, 100)
count = 1

if rdNumber < 1 and rdNumber > 100:
  print("Please choose the number between 1-100")
else:
  ansNumber = 0
  while ansNumber != rdNumber:
    ansNumber = int(input("What do you think the number is?: "))
    if ansNumber > rdNumber:
      print("Wrong, your answer is greater than the correct number.")
      count += 1
    elif ansNumber < rdNumber:
      print("Wrong, your answer is less than the correct number.")
      count += 1
if count > 1:
  print("Correct!!! You've guessed for ", count, " times")
else:
  print("Correct!!! You've guessed for ", count, " time")
```
